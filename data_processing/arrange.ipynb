```python
import os
import re
import pandas as pd 
import openpyxl

xname = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', f"{'xlsx_#3_columnize'}.xlsx")
df = pd.read_excel(xname, index_col=0)

def find_name(job_list):
    raw_data = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', "{}.xlsx".format(job_list))
    wb = openpyxl.load_workbook(raw_data) 
    ws = wb.worksheets[0] 

    mr = ws.max_row 

    result = []
    filtrate = re.compile(u'[\u4e00-\u9fff]')
    
    for i in range (2, mr + 1):
        r_list = []
        name = ws.cell(row = i, column = 2)
        info = ws.cell(row = i, column = 3)
        
        p_name = str(name.value).replace('(', '').replace(')', '')
        p_name = filtrate.sub(r'', p_name)
        p_info = str(info.value).split("/")
        
        r_list.append(p_name)
        for i in range(len(p_info)):
            r_list.append(p_info[i])
        
        result.append(r_list)
    
    return result

def phonor(df):
    h_list = find_name('honor_list')

    H1_AGE, H1_SEX, H2_AGE, H2_SEX, H3_AGE, H3_SEX = list(), list(), list(), list(), list(), list()

    for i in range(len(df)):
        data = df.iloc[i, 11].split()
        h1, h2, h3 = str(), str(), str()
        
        h1_age, h1_sex, h2_age, h2_sex, h3_age, h3_sex = int(), int(), int(), int(), int(), int()

        for j in range(len(data)):
            if len(h3)>0:
                break
            if '판사' in data[j] and len(h2)>0:
                h3 = data[j+1].replace("(", "").replace(")", "").replace("]", "").replace("[", "").replace("'", "").replace(",", "")
            elif '판사' in data[j] and len(h1)>0:
                h2 = data[j+1].replace("(", "").replace(")", "").replace("]", "").replace("[", "").replace("'", "").replace(",", "")
            elif '판사' in data[j]:
                if j == len(data)-1:
                    break
                h1 = data[j+1].replace("(", "").replace(")", "").replace("]", "").replace("[", "").replace("'", "").replace(",", "")

        for j in range(len(h_list)):
            if len(h1)>0 and h1_age == 0 and h1 in h_list[j][0]:
                for k in range(1, len(h_list[j])):
                    if '년생' in h_list[j][k]:
                        h1_age = h_list[j][k].replace('년생', '')
                    if '여자' in h_list[j][k]:
                        h1_sex = 1
                    elif '남자' in h_list[j][k]:
                        h1_sex = 2
                        
            if len(h2)>0 and h2 in h_list[j][0]:
                for k in range(1, len(h_list[j])):
                    if '년생' in h_list[j][k]:
                        h2_age = h_list[j][k].replace('년생', '')
                    if '여자' in h_list[j][k]:
                        h2_female = 1
                    elif '남자' in h_list[j][k]:
                        h2_male = 2

            if len(h3)>0 and h3 in h_list[j][0]:
                for k in range(1, len(h_list[j])):
                    if '년생' in h_list[j][k]:
                        h3_age = h_list[j][k].replace('년생', '')
                    if '여자' in h_list[j][k]:
                        h3_female = 1
                    elif '남자' in h_list[j][k]:
                        h3_male = 2
        
        H1_AGE.append(h1_age)
        H1_SEX.append(h1_sex)
        
        H2_AGE.append(h2_age)
        H2_SEX.append(h2_sex)

        H3_AGE.append(h3_age)
        H3_SEX.append(h3_sex)
        
    df['H1_AGE'] = H1_AGE
    df['H1_SEX'] = H1_SEX

    df['H2_AGE'] = H2_AGE
    df['H2_SEX'] = H2_SEX

    df['H3_AGE'] = H3_AGE
    df['H3_SEX'] = H3_SEX

    return df


def pprosec(df):
    
    p_list = find_name('prosecutor_list')

    P1_AGE, P1_SEX, P2_AGE, P2_SEX = list(), list(), list(), list()

    for i in range(len(df)):
        data = df.iloc[i, 6].split()
        p1, p2 = str(), str()
        p1_age, p1_sex, p2_age, p2_sex = int(), int(), int(), int()

        for j in range(len(data)):
            if '기소' in data[j]:
                p1 = data[j].replace('기소', '').replace("(", "").replace(")", "").replace("[", "").replace("'", "").replace(",", "")
            elif '공판' in data[j]:
                p2 = data[j].replace('공판', '').replace("(", "").replace(")", "").replace("]", "").replace("'", "").replace(",", "")

        for j in range(len(p_list)):
            if p1 in p_list[j][0]:
                for k in range(1, len(p_list[j])):
                    if '년생' in p_list[j][k]:
                        p1_age = p_list[j][k].replace('년생', '')
                    if '여자' in p_list[j][k]:
                        p1_female = 1
                    elif '남자' in p_list[j][k]:
                        p1_male = 2
                        
            if p2 in p_list[j][0]:
                for k in range(1, len(p_list[j])):
                    if '년생' in p_list[j][k]:
                        p2_age = p_list[j][k].replace('년생', '')
                    if '여자' in p_list[j][k]:
                        p2_female = 1
                    elif '남자' in p_list[j][k]:
                        p2_male = 2

        P1_AGE.append(p1_age)
        P1_SEX.append(p1_sex)
        
        P2_AGE.append(p2_age)
        P2_SEX.append(p2_sex)
        
    df['P1_AGE'] = P1_AGE
    df['P1_SEX'] = P1_SEX
    df['P2_AGE'] = P2_AGE
    df['P2_SEX'] = P2_SEX

    return df


df1 = pprosec(df)
df2 = phonor(df1)
```
