```python
import pdftotext
import os

def text22text(finput):
    
    fout = os.path.join(os.getcwd(), 'output.txt')
    
    if '.pdf' in finput:
         
        with open(finput, "rb") as f:
            pdf = pdftotext.PDF(f)

        with open('output.txt', 'w',-1,'utf-8') as f:
            f.write("\n\n".join(pdf))
    
        with open('output.txt', 'r',-1,'utf-8') as f:
            text=f.read()
            
    elif '.HWP' or '.hwp' in finput:
        
        hwp2txt_cmd = r'hwp5txt --output={} {}'.format(fout, finput)
        os.system(hwp2txt_cmd)
        with open(fout,'r',encoding='UTF8') as f:
            text=f.read()
            
    return text
```

```python
import re
text = text22text("(2015. 9. 18.자 형사) 마사지 업소의 업무를 방해하고, 직원에게 상해를 가한 피고인에 대하여 실형을 선고한 사례.pdf")
#print(text)
text = re.sub(r"(-\s?\d\s?-)", ' ', text)
text = re.sub(r' +', ' ', text)
text = re.sub(r'_{2,}', '', text)
case = re.search(r'\b사 건\b(.+)', text).group(1)
undertrial = re.search(r'\b피 고 인\s(.+)', text).group(1)
prosecutor = re.search(r'\b검 사\s((.+),(.+))', text).group(1)
date = re.search(r'\b판 결 선 고\s(.+)', text).group(1)
sentence = re.search(r'\b주 문\b\n{2,}(.+)', text).group(1)
if '재판장' in text:
    honor = re.findall(r'\b재판장|판사 \b(.+)', text)
    body = text[text.index('이 유'):text.rindex('재판장')]
else:
    honor = re.search(r'\b판사 \b(.+)', text).group(1)
    body = text[text.index('이 유'):text.rindex('판사')]
    
body = body.replace('이 유', '').replace('범 죄 사 실', '').replace('증거의 요지', '').replace('법령의 적용', '').replace('양형의 이유', '')
body = re.sub("\n{2,}", "", body)
body = re.sub(r'(\d+\.\s\d+\.\s\d+\.)(\s\d+:\d+경)?', '', body)

print(case, '\n', undertrial, '\n', prosecutor, '\n', date, '\n', sentence, '\n', honor)
# 제53조, 제55조 제1항 제3호(
# () [] 
```
