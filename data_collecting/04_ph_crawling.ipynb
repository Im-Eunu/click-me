```python
import os
import time
import pandas as pd
from selenium import webdriver

driver = webdriver.Chrome("./chromedriver")

driver.get("https://www.lawnb.com/")

time.sleep(1)
driver.find_element_by_css_selector("dl dd:nth-of-type(5) a").click()
time.sleep(1)
driver.find_element_by_css_selector("div.submenu_group dl:nth-of-type(5) dd a").click()
time.sleep(1)
driver.find_element_by_css_selector("select#cJobCode option:nth-of-type(4)").click()
time.sleep(1)
driver.find_element_by_css_selector("input#searchBtn").click()

result_list=[]
new_dict=[]

for i in range(1, 1031):
    
    
    page_bar = driver.find_elements_by_css_selector("div.navigationLinks > *")
    
    time.sleep(12)
    legal_profession = driver.find_elements_by_css_selector("div#search_results li")
    
    
    for legal in legal_profession:
        name = legal.find_element_by_css_selector("h3 a").text
        info = legal.find_element_by_css_selector("div h3 span span").text
        new_dict={
            'name': name,
            'info': info,
        }
        result_list.append(new_dict)
    
    time.sleep(2)
    page_bar = driver.find_elements_by_css_selector("div.navigationLinks > *")
    time.sleep(2)
    
    try:
        if i < 6:
            if i%5 != 0:
                page_bar[i%5].click()
            else:
                page_bar[5].click()
        else:
            if i%5 != 0:
                page_bar[i%5+2].click()
            else:
                page_bar[7].click()
    except:
        print("수집완료")
        break

        
df = pd.DataFrame(result_list)  
txt_name = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', f"{'attorney_list'}.xlsx")
excel_writer = pd.ExcelWriter(txt_name, engine='xlsxwriter')
df.to_excel(excel_writer)
excel_writer.save()
```
